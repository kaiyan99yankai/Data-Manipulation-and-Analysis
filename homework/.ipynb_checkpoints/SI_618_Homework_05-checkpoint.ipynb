## 4. Report the type-token ratio for each of the candidates (Trump & Biden), based on pooling the data from the two debates, and interpret the differences between the two candidates.

In text analysis, a 'type' is a unique word, whereas a 'token' is any word.  For example, the sentence

"One fish, two fish, red fish, blue fish"

contains 8 tokens, but 5 types (one, two, red, blue, fish).

Interestingly, the type-token ratio can be used to monitor changes in the use of vocabulary items in children with under-developed vocabulary and/or word finding difficulties and, for example, in adults who have suffered a stroke and who consequently exhibit word retrieval difficulties and naming difficulties. [Ref](https://www.sltinfo.com/type-token-ratio/)

````python
trump_token = []
trump_type = set()
for i in range(len(debate_1[debate_1.speaker=='President Donald J. Trump'])):
    doc = nlp(debate_1[debate_1.speaker=='President Donald J. Trump'].text.iloc[i])
    for token in doc:
        trump_token.append(token.text)
        trump_type.add(token.text)
for i in range(len(debate_2[debate_2.speaker=='Donald Trump'])):
    doc = nlp(debate_2[debate_2.speaker=='Donald Trump'].text.iloc[i])
    for token in doc:
        trump_token.append(token.text)
        trump_type.add(token.text)
print(len(trump_type)/len(trump_token))
````

````python
biden_token = []
biden_type = set()
for i in range(len(debate_1[debate_1.speaker=='Vice President Joe Biden'])):
    doc = nlp(debate_1[debate_1.speaker=='Vice President Joe Biden'].text.iloc[i])
    for token in doc:
        biden_token.append(token.text)
        biden_type.add(token.text)
for i in range(len(debate_2[debate_2.speaker=='Joe Biden'])):
    doc = nlp(debate_2[debate_2.speaker=='Joe Biden'].text.iloc[i])
    for token in doc:
        biden_token.append(token.text)
        biden_type.add(token.text)
print(len(biden_type)/len(biden_token))
````

From the above data, we find that the type-ratio rate of Trump and Biden are respectively around 0.106 and 0.127. Biden has a slight higher rate than Trump. The type ratio rate may reflect how developed one's vocabulary is. From the result, we may think Biden has more vocalbuary than Trump, who tends to speak more in a repetative way.

````python
import nltk
nltk.download('vader_lexicon')
````

````python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent_analyzer = SentimentIntensityAnalyzer()
def sentiment_scores(docx):
    return sent_analyzer.polarity_scores(docx.text)
````

````python
from spacy.tokens import Doc
Doc.set_extension("sentimenter",getter=sentiment_scores)
````

````python
debate_2['rating'] = debate_2['text'].apply(lambda x: nlp(x)._.sentimenter['compound'])
debate_2
````

````python
debate_2.groupby('speaker').mean()
````

````python
hour_list = [0]
for i in range(1, len(debate_2)):
    if int(debate_2.minute.iloc[i][:2])<int(debate_2.minute.iloc[i-1][:2]):
        hour_list.append(hour_list[i-1]+1)
    else:
        hour_list.append(hour_list[i-1])
debate_2['hour'] = hour_list
debate_2.hour = debate_2.hour.astype(str)
````

````python
debate_2_new = debate_2
debate_2_new.minute = '0'+debate_2.hour+':'+debate_2_new.minute
debate_2_new.minute
````

````python
debate_2_trump = debate_2_new[debate_2_new.speaker=='Donald Trump']
debate_2_trump['timestamp'] = pd.to_timedelta(debate_2_trump.minute)
plt.plot(debate_2_trump.set_index('timestamp').rolling('300s').mean()['rating'])
````
